```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sd_data=pd.read_csv("C:\\Users\\kulka\\Documents\\PDS\\StudentPerformance\\Raw_Data\\cleaned_student_data.csv")
```

```python
sd_data.head()
```

```python
data1 = sd_data.pivot_table(index='parental level of education', values=['math score', 'reading score', 'writing score'], aggfunc='mean')

data1.plot(kind='bar', figsize=(14, 8), colormap='viridis')

plt.title("Student scores and Parental Level of Education Relation")
plt.xlabel("Parental Level of Education")
plt.ylabel("Scores")
plt.legend(title="Score Type", loc="upper right")
plt.tight_layout()
plt.show()
```

```python
#Distribution of male and female students
gender_counts = sd_data['gender'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['lightblue', 'lightcoral'])
plt.title('Distribution of Male and Female Students')
plt.show()
```

```python
data3 = sd_data.groupby('test preparation course')[['math score', 'reading score', 'writing score']].mean()
print(data3)
# Create a bar plot
data3.plot(kind='bar', figsize=(10, 6), colormap='viridis')

plt.title("Student Scores and Test Preparation Course Relation")
plt.xlabel("Test Preparation Course")
plt.ylabel("Average Scores")
plt.legend(title="Score Type", loc="upper right")
plt.xticks(rotation=0)  # Rotate x-axis labels if needed
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

plt.scatter(sd_data[sd_data['gender'] == 'male']['gender'], 
            sd_data[sd_data['gender'] == 'male']['writing score'], 
            color='blue', label='Male', alpha=0.7)

# Scatter plot for female students
plt.scatter(sd_data[sd_data['gender'] == 'female']['gender'], 
            sd_data[sd_data['gender'] == 'female']['writing score'], 
            color='red', label='Female', alpha=0.7)

plt.title('Gender vs. Math Scores')
plt.xlabel('Gender')
plt.ylabel('Math Scores')
plt.xticks([0, 1], ['Male', 'Female'])
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
data4 = pd.crosstab(sd_data['race/ethnicity'], sd_data['lunch'])

# Create a grouped bar plot
data4.plot(kind='bar', figsize=(12, 6), colormap='viridis')

plt.title("Distribution of Lunch Type by Race/Ethnicity")
plt.xlabel("Race/Ethnicity")
plt.ylabel("Count")
plt.legend(title="Lunch Type", loc="upper right")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
